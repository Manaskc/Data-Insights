```python
import pandas as pd

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
df = pd.read_csv('C:\\Users\\manasan\\Desktop\\New folder\\NSE-Tata-Global-Beverages-Limited1.csv')
print(df)
```

```python
x = df[['High','Open','Low','Total Trade Quantity']].values
y = df['Close'].values
```

```python
plt.figure(figsize=(16,8))
plt.plot(df["Close"],label='Close Price history')
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
print(regressor.coef_)
```

```python
predict = regressor.predict(x_test)
result = pd.DataFrame({'Actual':y_test.flatten(),'Predicted':predict.flatten()})
result.head(25)
```

```python
plt.figure(figsize=(16,8))
plt.plot(result['Predicted'],'b',label="predicted value",linewidth=2)
plt.plot(result['Actual'],'r',label="Actual value",linewidth=2)
plt.title("comparison between predicted and Actual")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.legend()
plt.grid(True,color='k')
plt.show();
```
